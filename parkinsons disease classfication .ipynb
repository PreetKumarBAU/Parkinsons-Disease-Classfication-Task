```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
dataset =pd.read_csv(r"C:\Users\90536\Downloads\parkinsons.csv")

dataset.head()
```

```python
j = list(range(1,24))
print(j)
j.remove(17)
print(j)
```

```python
#Independent Variables
X = dataset.iloc[ : ,j  ]
X
```

```python
#Dependent variable 
Y = dataset.iloc[: , 17]
Y
```

```python
#Split the DATASET into test and train set
from sklearn.model_selection import train_test_split

trainX , testX , trainY , testY = train_test_split(X , Y , test_size = 0.2 , random_state = 0)
```

```python
trainX.head()
```

```python
#Feature Scaling
#So that faster computation/learning of Model
#All features are SAME SCALED so all features are considered equally 
#As machine learning Models consider EUCLIDEAN DISTANCE while learning . Thus all features must be same scaled
from sklearn.preprocessing import StandardScaler

SS = StandardScaler()

X_train  = SS.fit_transform(trainX)
X_test = SS.transform(testX)

```

```python
#Dimension Reduction for selecting LESS FEATURES and ONLY IMPORTANT Features

from sklearn.decomposition import PCA

pca = PCA(n_components = None)

X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)

variance = pca.explained_variance_ratio_
variance
```

```python
#Selected Only two features as their variance explained is 59% and 10% whereas third feature only explained 0.07 variance so remove further features
pca = PCA(n_components = 2)

X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

```python
#Applying Different Machine learning Models

#Knn Model
from sklearn.neighbors import KNeighborsClassifier
#Calling the Method and setting the parameters
knnModel = KNeighborsClassifier(n_neighbors = 8 , p =2 , metric = "minkowski")
#Training the  K nearest neighbor Model
knnModel.fit(X_train ,trainY )
```

```python
#Predicating the Output on Test Data

YPred = knnModel.predict(X_test)
YPred 
```

```python
#Finding Accuracy of the Model
from sklearn.metrics import accuracy_score

accuracy_score(testY ,YPred  )
```
